###항공통계 화면 출력하기   
####2021년 월마다 항공사 데이터 엑셀 다운로드 자동화   
####검색누르고 엑셀 다운로드하기

```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.select import Select
import time

driver = webdriver.Chrome()     
#Chromedriver을 사용하기 때문에 Chrom으로 변경

driver.get("https://www.airportal.go.kr/knowledge/statsnew/air/airline.jsp#")  

#시작년도 메뉴 선택
#dropdown = driver.find_element(By.CSS_SELECTOR,'.form-control').click()

driver.find_element(By.XPATH, "//*[@id='start_year']/option[text()='2021']").click()
driver.find_element(By.XPATH, "//*[@id='end_year']/option[text()='2021']").click()

#1월부터 12월까지 for문으로 클릭되도록 구현.
for i in range(1,13):
    driver.find_element(By.XPATH, "//*[@id='start_month']/option[text()="+str(i)+"]").click()
    driver.find_element(By.XPATH, "//*[@id='end_month']/option[text()="+str(i)+"]").click()
    driver.find_element(By.CSS_SELECTOR,'.mainSearchBtn').click()  #검색버튼 누르기
    time.sleep(2) #월->검색버튼을 누른뒤 2초뒤에 엑셀다운로드되도록

    driver.find_element(By.CSS_SELECTOR,'.mainExcellBtn').click()   #엑셀다운받기누르기
    time.sleep(3)


driver.close()
```

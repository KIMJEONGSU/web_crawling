#웹 크롤링 + 파일 저장   
##벅스홈페이지에서 음악 순위가져오기

```python
from urllib.request import urlopen                # 웹 페이지의 HTML을 가져오는 모듈
from bs4 import BeautifulSoup    # HTML을 파싱하는 모듈
import pandas as pd
 
url=urlopen('https://music.bugs.co.kr/chart?wl_ref=M_left_02_01%20%EC%B6%9C%EC%B2%98:%20https://seungjuitmemo.tistory.com/203%20[Memo%20Memo:%ED%8B%B0%EC%8A%A4%ED%86%A0%EB%A6%AC]')

soup=BeautifulSoup(url.read(),'html.parser')


musics = soup.find_all('td',{'class':'check'})
f = open('C:/Users/************/새파일.txt', 'w')

for i, music in enumerate(musics):
    data=("{}위: {}".format(i+1, music.input['title'])+'\n')
    f.write(data)
f.close()    
    
```
